# Question 10

- Given X matrix below, 
- Find the column with maximum column sum
- For which rows of the matrix, the sum of the first three elements of the row is greater than the sum of the last two elements of the row. We can use `np.where`.

````python
np.random.seed(10)
X = np.random.rand(5,5)
X
````

````python
a = np.sum(X, axis=0) # 각 column의 sum을 계산 후 배열에 저장
idx = np.argmax(a) # column sum이 max인 원소의 index를 return
idx # the column with maximum column sum is column 3
````

````python
XLeft = X[:, 0:3] # matrix of first three elements
XRight = X[:, 3:5] # matrix of last two elements
#resLeft = np.sum(XLeft, axis=1)
#resRight = np.sum(XRight, axis=1)
#print(resLeft)
#print(resRight)
idx = np.where(np.sum(XLeft, axis=1) > np.sum(XRight, axis=1)) #왼쪽 > 오른쪽인 row의 index만 추출
print(idx[0].tolist()) #np.where의 결과는 tuple, list로 변환해서 구하고자 하는 row의 index를 list로 출력
# rows that the sum of first three elements of the row
# is greater than the sum of the last two elements of the row
# are row 0, 1, 2
````
